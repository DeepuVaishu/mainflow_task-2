```python
data2 = data1.drop(['video_id','publish_time','tags','thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description','trending_date','title', 'channel_title', 'category_id', 'description'], axis=1)
data2.head()
```

```python
Q1 = data2.quantile(0.25)
Q3 = data2.quantile(0.75)
IQR = Q3 - Q1   
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data2[~((data2 < lower_bound) | (data2 > upper_bound)).any(axis=1)]
outliers
```
